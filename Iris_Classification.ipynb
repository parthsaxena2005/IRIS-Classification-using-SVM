# IRIS Classification Project using SVM

### Importing the IRIS dataset

```python
from sklearn import datasets
iris = datasets.load_iris()
description = iris.DESCR
print(description)
```

### Importing the necessary libraries

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

### Creating test and train data

here we have used 70% of the data to train the models and the rest 30% to test them
the distribution is startifird so that models which are heavily biased towards a outcome are not obtained

```python
X = pd.DataFrame(iris.data)

#X = iris.data
Y = pd.DataFrame(iris.target)
print(X)
print(Y)


X_test , X_train, Y_test, Y_train = train_test_split(X,Y, test_size=0.3, random_state=1234, stratify=Y)
```

## SVM Kernels
we have options for multiple SVM Kernels with different working and each one suitable for different type of data

Thus here we will be testing our data to select the best kernel for our need, and make predictions using it

### RBF Kernels
these require a value for the gamma factor, so we will try 3 different values

using gamma = 1.0

```python
rbf_kernel  = SVC(kernel='rbf', gamma=1.0)
rbf_kernel.fit(X_train, Y_train)
Y_predict = rbf_kernel.predict(X_test)
cmf_rbd = confusion_matrix(Y_test, Y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cmf_rbd)
disp.plot()

score_rbf = rbf_kernel.score(X_test, Y_test)
print("rbf score = ")
print(score_rbf)
```

using gamma = 0.5

```python
rbf_kernel  = SVC(kernel='rbf', gamma=0.5)
rbf_kernel.fit(X_train, Y_train)
Y_predict = rbf_kernel.predict(X_test)
cmf_rbd = confusion_matrix(Y_test, Y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cmf_rbd)
disp.plot()

score_rbf = rbf_kernel.score(X_test, Y_test)
print("rbf score = ")
print(score_rbf)
```

using gamma = 5.0

```python
rbf_kernel  = SVC(kernel='rbf', gamma=5.0)
rbf_kernel.fit(X_train, Y_train)
Y_predict = rbf_kernel.predict(X_test)
cmf_rbd = confusion_matrix(Y_test, Y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cmf_rbd)
disp.plot()

score_rbf = rbf_kernel.score(X_test, Y_test)
print("rbf score = ")
print(score_rbf)
```

### Linear Kernel

```python
linear_kernel  = SVC(kernel='linear')
linear_kernel.fit(X_test, Y_test)
Y_predict_linear = linear_kernel.predict(X_test)
cmf_linear = confusion_matrix(Y_test, Y_predict_linear)
disp = ConfusionMatrixDisplay(confusion_matrix=cmf_linear)
disp.plot()

score_linear = linear_kernel.score(X_test, Y_test)
print("Linear score = ")
print(score_linear)
```

### Sigmoid Kernel

```python
sigmoidal_kernel  = SVC(kernel='sigmoid')
sigmoidal_kernel.fit(X_test, Y_test)
Y_predict_sigmoid = sigmoidal_kernel.predict(X_test)
cmf_sig = confusion_matrix(Y_test, Y_predict_sigmoid)
disp = ConfusionMatrixDisplay(confusion_matrix=cmf_sig)
disp.plot()

score_sig = sigmoidal_kernel.score(X_test, Y_test)
print("Sigmoidal score = ")
print(score_sig)
```

### Polynomial Kernel

```python
polynomial_kernel  = SVC(kernel='poly')
polynomial_kernel.fit(X_test, Y_test)
Y_predict_poly = polynomial_kernel.predict(X_test)
cmf_poly = confusion_matrix(Y_test, Y_predict_poly)
disp = ConfusionMatrixDisplay(confusion_matrix=cmf_poly)
disp.plot()

score_poly = polynomial_kernel.score(X_test, Y_test)
print("Polynomial score = ")
print(score_poly)
```

## Observations
comparing accuracy scores and confusion matrices from all the kernels

rbf score (gamma=5.0)   = 0.9714285714285714

Linear score            = 0.9904761904761905

Sigmoidal score         = 0.10476190476190476

Polynomial score        = 0.9904761904761905

clearly the Sigmoidal SVM is not a good fit for out usecase

The Linear, Polynomial working best.

Saving the predictions

```python
predictions = pd.DataFrame()
predictions['result'] = Y_predict_poly
predictions.to_csv("Predictions.csv", header=True, index =False)
IrisClassificationData = pd.DataFrame(X)
IrisClassificationData.columns = iris.feature_names
IrisClassificationData.to_csv("Iris Classification Data.csv",header=True, index =False)
testdataset = pd.DataFrame(X_test)
testdataset.columns = iris.feature_names
testdataset.to_csv("Test_Dataset.csv",header=True, index =False)
testresult = pd.DataFrame()
testresult['result'] = Y_test
testresult.to_csv("Test_Result.csv",header=True, index =False)
```
